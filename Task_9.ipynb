## Sentiment Analysis

```python
# Model
mdlSnt = pipeline('sentiment-analysis')

# Parameters
#prmStatement = 'We are happy to go on vacation this spring break.'
prmStatement = ['It is nice to have you here in the US.  I heard the weather is terrible in the UK.',
        'I am struggling with stress']
# Predict
results=mdlSnt(prmStatement)
# Display results
for statement, result in zip(prmStatement, results):
    label = result['label']
    score = result['score']
    print(f"Statement: \"{statement}\"")
    print(f"Sentiment: {label} (Score: {score:.4f})\n")

```

Sentiment Analysis can be used in ordering choices.

```python
from transformers import pipeline

# Load the sentiment-analysis model
mdlSnt = pipeline('sentiment-analysis')

# List of statements
prmStatement = [
    'I am student in Bilkent University','I am student in Middle East Technical University','I am student in Koç University','I am student in Boğaziçi University','I am student in Oxford University','I am student in MIT','I am student in Sabancı University'
,'I am student in Abdullah Gül University']

# Get sentiment analysis results
results = mdlSnt(prmStatement)

# Combine statements with their results
combined_results = list(zip(prmStatement, results))

# Sort by sentiment score:
# - Positive sentiments: sorted by score in descending order
# - Negative sentiments: sorted by score in ascending order
combined_results_sorted = sorted(
    combined_results,
    key=lambda x: (x[1]['label'] == 'NEGATIVE', -x[1]['score']) if x[1]['label'] == 'POSITIVE' else (x[1]['label'] == 'NEGATIVE', x[1]['score'])
)

# Display sorted results with formatting
print("\nUNIVERSITY PREFERENCE:\n")
for index, (statement, result) in enumerate(combined_results_sorted, start=1):
    label = result['label']
    score = result['score']
    print(f"{index}. \"{statement}\"")
    print(f"   Sentiment: {label}")
    print(f"   Score: {score:.4f}\n")
```

Hmmmm, there are interesting results.

```python
from transformers import pipeline

# Load the sentiment-analysis model
mdlSnt = pipeline('sentiment-analysis')

# List of statements
prmStatement = [
    'I am mechanical engineer','I am industrial engineer','I am computer engineer','I am electrical and electronics engineer'
]

# Get sentiment analysis results
results = mdlSnt(prmStatement)

# Combine statements with their results
combined_results = list(zip(prmStatement, results))

# Sort by sentiment score:
# - Positive sentiments: sorted by score in descending order
# - Negative sentiments: sorted by score in ascending order
combined_results_sorted = sorted(
    combined_results,
    key=lambda x: (x[1]['label'] == 'NEGATIVE', -x[1]['score']) if x[1]['label'] == 'POSITIVE' else (x[1]['label'] == 'NEGATIVE', x[1]['score'])
)

# Display sorted results with formatting
print("\nENGINEERING DEPARTMENT PREFERENCE:\n")
for index, (statement, result) in enumerate(combined_results_sorted, start=1):
    label = result['label']
    score = result['score']
    print(f"{index}. \"{statement}\"")
    print(f"   Sentiment: {label}")
    print(f"   Score: {score:.4f}\n")
```

Can perform sentiment analysis in other languages as well using `model = 'nlptown/bert-base-multilingual-uncased-sentiment'`

```python
from transformers import pipeline

# Load the sentiment-analysis model
mdlSnt_Turkish = pipeline('sentiment-analysis', model='savasy/bert-base-turkish-sentiment-cased')

# List of statements
prmStatement_Turkish = [
    'Ben makine mühendisiyim','Ben endüstri mühendisiyim','Ben bilgisayar mühendisiyim','Ben elektrik ve elektronik mühendisiym'
]

# Get sentiment analysis results
results_Turkish = mdlSnt_Turkish(prmStatement_Turkish)

# Combine statements with their results
combined_results_Turkish = list(zip(prmStatement_Turkish, results_Turkish))

# Sort by sentiment score:
# - Positive sentiments: sorted by score in descending order
# - Negative sentiments: sorted by score in ascending order
combined_results_sorted_Turkish = sorted(
    combined_results_Turkish,
    key=lambda x: (x[1]['label'] == 'NEGATIVE', -x[1]['score']) if x[1]['label'] == 'POSITIVE' else (x[1]['label'] == 'NEGATIVE', x[1]['score'])
)

# Display sorted results with formatting
print("\nMühendislik departman Tercihi:\n")
for index, (statement, result) in enumerate(combined_results_sorted_Turkish, start=1):
    label = result['label']
    score = result['score']
    print(f"{index}. \"{statement}\"")
    print(f"   Sentiment: {label}")
    print(f"   Score: {score:.4f}\n")
```

It also works well in Turkish but interestingly they are all described as a negative and have different score from English statments.

```python
#context = Henry Ford's Bioraphy
question = "When was the Ford Motor Company Founded?"

# Predict
mdlQa(question = question, context = context)
```

It is not correct. The answer is birtday date of Henry Ford, not Ford Motor Company foundation date.The model takes <font color='#378'>"Henry Ford, founder of Ford Motor Company, was born in Springwells Township, Wayne County, Michigan, on July 30, 1863."<font color='#999'> as an answer instead of <font color='#472'>"On June 16, 1903, Henry and 12 others invested $28,000 and created Ford Motor Company."

In first insight, the reason of mistake seems like pickin up closest data behavior of the model but when we change the place of these above two phrases, the answer does not change. It may be solved by adding more context into model.

```python
#context = Henry Ford's Bioraphy
question = "When was the Ford Motor Company Founded by twelve investers?"

# Predict
mdlQa(question = question, context = context)
```

```python
#context = Henry Ford's Bioraphy
question = "When was the Ford Motor Company Founded by 12 investers?"

# Predict
mdlQa(question = question, context = context)
```

While adding more detail to the question gives still inaccurate answer, adding key word '12' gives accurate result. Hence, it can be claimed that the effect of keyword on model is more than effect of more detail and more text.
